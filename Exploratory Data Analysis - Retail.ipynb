```python
# importing the necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# loading dataset

data = pd.read_csv("G:\SampleSuperstore.csv")
```

```python
# displaying the first five rows of dataset

data.head()
```

```python
# Description of dataset

data.describe()
```

```python
# All information of dataset

data.info()
```

```python
# Checking the dimensions of dataset

data.shape
```

```python
# Checking whether any empty or null value is there or not  

data.isnull().sum()
```

```python
# Displaying the columns of the dataset

data.columns
```

```python
# Checking number of unique values in each column

data.nunique()
```

```python
# Calculating the average profit so that we can check top 10 profitable countries

avg_profit = data['Profit'].astype('float').mean()
print("The average  profit is {}".format(avg_profit))
```

```python
# Checking the total sales and total profit

total_sales=data['Sales'].sum()
total_profit=data['Profit'].sum()
print(total_sales)
print(total_profit)
```

```python
# Displaying the max profit

data['Profit'].max()
```

```python
# Grouping top 10 sales state according to the our data

Top_10_Sales = data.groupby("State").Sales.sum().nlargest(n =10)
Top_10_Profits = data.groupby("State").Profit.sum().nlargest(n =10)
```

```python
# Displaying top sales states

Top_10_Sales
```

```python
# Displaying top 10 profit state

Top_10_Profits
```

```python
# Plotting the top 10 states by sales

plt.style.use('seaborn')
Top_10_Sales.plot(kind ='bar', figsize =(10,10), fontsize =14,color= 'red')
plt.xlabel("States", fontsize =13)
plt.ylabel("Total Sales",fontsize =13)
plt.title("Top 10 States by Sales",fontsize =16)
plt.show()
```

```python
# Plotting the top 10 states by profit

plt.style.use('seaborn')
Top_10_Profits.plot(kind ='bar', figsize =(10,10), fontsize =14,color='green')
plt.xlabel("States", fontsize =13)
plt.ylabel("Total profit",fontsize =13)
plt.title("Top 10 States by Profit",fontsize =16)
plt.show()
```

# Seaborn library is mainly used for data visualization based on matplot

```python
sns.heatmap(data.corr(), annot = True)
```

```python
sns.lmplot(x='Category',y='Sales',data=data,fit_reg=False,hue='Sub-Category',legend=True,palette='winter',size=10)
plt.title("Sales per Category and Sub-Category",fontsize=30)
```

From the above graph we can see that Technology has more sales in it.

```python
PL = pd.Series([],dtype=pd.StringDtype())
for i in range(len(data)):
    if data['Profit'][i] > 0:
        PL[i] = 'Profit'
    else:
        PL[i] = 'Loss'

        
data.insert(loc = 11, column = 'PL', value = PL)
```

```python
sns.countplot('PL', data=data)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='PL', y = 'Sales', data = data, hue='Category')
```

We can infer from the above plot, technology products can give us huge profit as well as huge losses at times. ON the other hand hand in the office supplies, the profit is lower but the loss is also very low that means the risk of doing business is less. Furniture products have a bit higher loss number than profit.

# Plotting the sales by sub-category

```python
plt.figure(figsize=(13,8))
sns.set(style="whitegrid", color_codes=True)
pal = sns.color_palette("colorblind", len(data))
sns.barplot(x='PL', y = 'Sales', data = data, palette = pal, hue='Sub-Category')
```

# Discount effect on Profit and Sales of Product

```python
plt.figure(figsize=(13,10))

pal = sns.color_palette("colorblind", len(data))
plt.subplot(2,1,1)
sns.barplot(x='Discount', y='Profit', palette = pal, data=data)

plt.subplot(2,1,2)
sns.barplot(x='Discount', y='Sales', palette = pal, data=data)
```

# Profits by Region and States

```python
plt.figure(figsize=(16,8))

pal = sns.color_palette("colorblind", len(data))
plt.subplot(2,1,1)
sns.barplot(x='Region', y='Profit', palette = pal, data=data)
```

We can see, the shops of west region makes the highest profit and the shops of east comes after that. The company needs to focus on the profit of Central region shops as its number is very low.

```python
plt.figure(figsize=(18,8))

pal = sns.color_palette("colorblind", len(data))

sns.barplot(x='State', y='Profit', palette = pal, data=data)
plt.xticks(rotation=60)
```

We can see around 10 states arein losses and others are gaining profits

```python
plt.figure(figsize=(12,8))

pal = sns.color_palette("colorblind", len(data))

sns.countplot('Ship Mode', palette = pal, data=data)
```

Standard class is the most popular shipping mode.
